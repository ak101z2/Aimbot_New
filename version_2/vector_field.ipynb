```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
TABLE_WIDTH = 36 
TABLE_LENGTH = 84 

TARGET_X = 9
TARGET_Y = 42

PUCK_WIDTH = 2 
PUCK_LENGTH = 2 

GRADIENT_WIDTH = TABLE_WIDTH / PUCK_WIDTH
GRADIENT_LENGTH = TABLE_LENGTH / PUCK_LENGTH
```

```python
# Meshgrid
x, y = np.meshgrid(np.linspace(0, GRADIENT_WIDTH, 10), 
                   np.linspace(0, GRADIENT_LENGTH, 10))

# Directional vectors
u = (np.full((10, 10), TARGET_X) - x) 
v = (np.full((10, 10), TARGET_Y) - y)

u = u / np.sqrt(u**2 + v**2)
v = v / np.sqrt(u**2 + v**2)

# Plotting Vector Field with QUIVER
plt.quiver(x, y, u, v, color='g')
plt.title('Vector Field')
  
# Setting x, y boundary limits
plt.xlim(0, GRADIENT_WIDTH)
plt.ylim(0, GRADIENT_LENGTH)
  
# Show plot with grid
plt.grid()
plt.show()
```
